```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(-0.1 * x) * np.sin(3 * x)
y4 = x**2

# Create a 2x2 subplot layout
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# Plot in the first subplot
axes[0, 0].plot(x, y1, color='blue', label='y1')
axes[0, 0].set_ylabel('y1', color='blue')
axes[0, 0].tick_params(axis='y', labelcolor='blue')

# Create a secondary y-axis for the first subplot
axes2_0 = axes[0, 0].twinx()
axes2_0.plot(x, y2, color='green', label='y2')
axes2_0.set_ylabel('y2', color='green')
axes2_0.tick_params(axis='y', labelcolor='green')

# Plot in the second subplot
axes[0, 1].plot(x, y3, color='red', label='y3')
axes[0, 1].set_ylabel('y3', color='red')
axes[0, 1].tick_params(axis='y', labelcolor='red')

# Create a secondary y-axis for the second subplot
axes2_1 = axes[0, 1].twinx()
axes2_1.plot(x, y4, color='purple', label='y4')
axes2_1.set_ylabel('y4', color='purple')
axes2_1.tick_params(axis='y', labelcolor='purple')

# Set titles and labels for the subplots
for ax in axes.flat:
    ax.set_xlabel('X-axis')
    ax.legend(loc='upper left')

# Adjust layout to prevent overlapping titles
plt.tight_layout()

# Show the plot
plt.show()
```
